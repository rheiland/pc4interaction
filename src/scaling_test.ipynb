```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
counts = []
times = [] 

def add_datum( counts, times , cnew, tnew ):
    counts.append(cnew);
    times.append(tnew); 
    return (counts,times); 

def add(cnew,tnew):
    return add_datum(counts,times,cnew,tnew);
```

```python
# Cost when we tell it not to test cells with zero volume 

add(998,21.40 )
add(1172, 22.1)
add(1546,23.3)
add(1961,29.5)
add(2342,34.7 )
add(3326,54.3)
add(5614,2*60+38.3)

plt.plot( counts, times , 'b-o')

coeff = np.polyfit( counts, times, 2 )
print( coeff )

cc = np.linspace( 0 , 10000 , 501)
tt = np.polyval( coeff, cc )

plt.figure(1)
plt.plot( counts, times , 'b-o')
cc = np.linspace( 0 , 10000 , 501)
tt = np.polyval( coeff, cc )
plt.plot( cc , tt , 'r')

plt.figure(2)
plt.plot( counts, times , 'b-o')
cc = np.linspace( 0 , 10000 , 501)
tt = np.polyval( coeff, cc )
plt.semilogy( cc , tt , 'r')
```
